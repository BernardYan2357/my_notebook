#### ç°åº¦å›¾è½¬æ¢

å¯¹BGRé€šé“å›¾åƒåšå¹³æ–¹å’ŒåŠ æƒå¹³å‡

ç°åº¦å›¾æè¿°äº†å›¾æ¡ˆçš„æ˜æš—åˆ†å¸ƒï¼ŒCMOSé‡‡é›†çš„å›¾åƒçš„ç°åº¦å›¾ä¹Ÿå¯ä»¥è®¤ä¸ºæ˜¯CMOSä¼ æ„Ÿå™¨æ¥æ”¶å…‰å­æ•°çš„åˆ†å¸ƒå›¾

````python
image=cv2.imread("./images/lena.png")
gray=cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
print(gray.shape)
imshow(gray=gray)
````

#### å›¾åƒæ‹¼æ¥

ä½¿ç”¨`np.hstack()`å’Œ`np.vstack()`å‡½æ•°æ°´å¹³å’Œå‚ç›´æ‹¼æ¥å›¾åƒ

æ¡ä»¶ï¼šæ‹¼æ¥çš„å›¾åƒå°ºå¯¸å¿…é¡»ç›¸åŒ

å¦‚æœå°ºå¯¸ä¸åŒï¼Œå¯ä»¥ä½¿ç”¨`cv2.resize()`å‡½æ•°è°ƒæ•´å›¾åƒå¤§å°

````python
opening=cv2.imread("./images/opening.png")
closing=cv2.imread("./images/closing.png")

(ox,oy,_)=opening.shape
(cx,cy,_)=closing.shape
closing_resized = cv2.resize(closing, (oy, ox),1,oy/cy)
vstack_img=np.vstack((image,image))
hstack_img=np.hstack((opening,closing_resized))

imshow(hstack_img=hstack_img,vstack_img=vstack_img)
````
